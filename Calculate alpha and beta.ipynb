```python
must_be_active_date = datetime.strptime('Jan 24 2020', '%b %d %Y')
def alpha_and_beta(index, baseline):
#     Inspired by https://blog.quantinsti.com/asset-beta-market-beta-python/
    index_data = pdr.get_data_yahoo(index,'2000-01-01','2030-01-01')
    baseline_data = pdr.get_data_yahoo(baseline,'2000-01-01','2030-01-01')

    if index_data.iloc[-1].name < must_be_active_date:
        raise "Shares no longer Active"
    
    return_index = index_data.Close.pct_change()[1:].values
    return_baseline = baseline_data.Close.pct_change()[1:].values
    
    minimum = min(len(return_index), len(return_baseline))
    minimum = min(minimum, 365 * 5)
    return_index = return_index[-minimum:]
    return_baseline = return_baseline[-minimum:]
    
    return_baseline = sm.add_constant(return_baseline)
    model = regression.linear_model.OLS(return_index, return_baseline).fit()
    return [minimum, model.params[0] * 365, model.params[1]]

alpha_and_beta('cqqq', 'spy')
```

```python
output = [['ticker', 'days vs spy', 'alpha vs spy', 'beta vs spy', 'days vs vgt', 'alpha vs vgt', 'beta vs vgt']]
for etf in ETFS:
    try:
        output.append([etf] + alpha_and_beta(etf, 'spy') + alpha_and_beta(etf, 'vgt'))
    except:
        print(etf)
```
